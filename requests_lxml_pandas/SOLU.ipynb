INTRODUCTION

1.1	Background

Worldwide large populations are exposed to high levels of air pollution. The amount of exposure in terms of intensity and time is far greater in the developing countries, causing various health impacts among the poorest and most vulnerable populations. In Nepal, population growth, rapid expansion of urban municipalities and increasing amounts of industrial and commercial activities has resulted in severe air quality issues (Das et al., 2018). Indoor Air pollution (IAP) from burning fuels for cooking also represents major environmental health burden. This problem is more severe in developing countries, where IAP ranks behind only malnutrition, unsafe water and sanitations as a major contributor to morbidity and mortality (Singh et al., 2012). IAP is the quality of air inside buildings as represented by concentrations of pollutants and thermal (temperature and relative humidity) conditions that affect the health, comfort and performance of occupants (Sharma, 2014).

Indoor pollution level of developing countries is approximately 100 times higher than the cities of the developed world with serious outdoor pollution but has received much attention from the researchers over the years (Dhakal, 2007). In addition to a lack of studies on the risks of exposures, there is a dearth of studies on the distribution and magnitude of exposure to IAP in developing countries itself (Smith 2002). The indoor air pollution studies conducted in Dolakha, Dang and Ilam establishes that the mud Improved Cooking Stove (ICS) is an appropriate intervention to reduce PM2.5 in rural kitchens and indicated positive changes in cough, phlegm, and eye irritation. This study recommends greater focus on proper operation and maintenance of ICS and ventilation to lower the smoke level even further (Singh et al., 2012). Such information is needed both to estimate the total burden of disease and to identify the regions and populations where interventions need to be focused.

2.1	Study Area
The study was carried out in Solududhkunda municipality of Solukhumbu district, which is one of 14 districts of Province No. 1 of eastern Nepal. Salleri is the district headquarters. The district consists of the sub-regions Solu and Khumbu covering an area of 3,312 km² and had a population 105,886 in 2011. Its highest elevation is 8848 m and lowest elevation is 600m.The district shares its border with Tibet to the North, Sankhuwasava and Bhojpur to the East, Okhaldhunga and Khotang to the South and Ramechap to the West. The district has 30 VDCs and 1 Municipality. Solududhkunda municipality is the only one municipality of Solukhumbu district formed by merging Salleri, Garma, Loding, Tamakhani, Beni, Takshindu, Kerung, Gorakhani and Tapting. The study covered the households of four wards of Solududhkunda municipality namely Solududhkunda- 1, Taksindu, Solududhkunda- 2, Beni, Solududhkunda- 3, Loding, Solududhkunda- 5, Salleri.
