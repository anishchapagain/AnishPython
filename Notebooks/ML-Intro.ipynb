# Regression
The term regression is used when you try to find the relationship between variables. In Machine Learning, and in statistical modeling, that relationship is used to predict the outcome of future events.

##Linear Regression --
Linear regression uses the relationship between the data-points to draw a straight line through all them. This line can be used to predict future values.

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html

```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]                  #y=m (slope) x + c (intercept)
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html#scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(age, speed)

#model predict!
def myfunc(age):
  return slope * age + intercept #speed

mymodel = list(map(myfunc, age)) #M,R,F

''''
This relationship - the coefficient of correlation - is called r.
The r value ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related.
'''
#print("MyModel :",mymodel[::25])
print("Slope: %f  \nIntercept: %f" % (slope, intercept))
print("R-squared: %f" % r_value**2) #high: 
print("R_Value: ",r_value)
print("P_Value: ",p_value)
print("Std_err: ",std_err)
```

```python
#The value of R-Squared is always between 0 to 1 (0% to 100%). r_value or correlation coefficient
#A high R-Squared value means that many data points are close to the linear regression function line.
#A low R-Squared value means that the linear regression function line does not fit the data well.

#The P-value is a statistical number to conclude if there is a relationship between..ages/speed.
#We test if the true value of the coefficient is equal to zero (no relationship). The statistical test for this is called Hypothesis testing.
#A common threshold of the P-value is 0.05 or 5%. If the P-value is lower than 0.05, we can reject the null hypothesis and conclude that it exist a relationship between the variables.

#Linear regression uses the least square method.
#The concept is to draw a line through all the plotted data points. The line is positioned in a way that it minimizes the distance to all of the data points.
#The distance is called "residuals" or "errors"

print(mymodel)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(age, speed)
plt.grid()
plt.plot(age, mymodel)
plt.show()
```

### Predict Future Values

```python
predict_speed = myfunc(15)
print(f"Age=15 > Speeed={predict_speed}")
predict_speed = myfunc(25)
print(f"Age=25 > Speeed={predict_speed}")
predict_speed = myfunc(35)
print(f"Age=35 > Speeed={predict_speed}")
predict_speed = myfunc(5)
print(f"Age=5 > Speeed={predict_speed}")
predict_speed = myfunc(1)
print(f"Age=1 > Speeed={predict_speed}")
```

## Machine Learning - Train/Test

'''
Train/Test is a method to measure the accuracy of your model. It is called Train/Test because you split the the data set into two sets: a training set and a testing set. 80% for training, and 20% for testing.

You train the model using the training set.You test the model using the testing set.
Train the model means create the model.Test the model means test the accuracy of the model.
'''

```python
x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(100, 40, 100) / x
plt.figure(figsize=(10,7))
plt.scatter(x, y)
plt.show()
```

```python
print(x[::20])
print(y[::20])
```

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

```python
plt.figure(figsize=(10,7))
plt.grid()
plt.scatter(train_x, train_y)
plt.scatter(test_x, test_y,color='r')
plt.show()
```

```python
#https://numpy.org/doc/stable/reference/generated/numpy.poly1d.html
#https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html?highlight=polyfit#numpy.polyfit     

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4)) #4: degree (numpy.polyfit: Least squares polynomial fit.)
from sklearn.metrics import r2_score

#Training
r2 = r2_score(train_y, mymodel(train_x)) #r_square: from sklearn!
print(r2)
```

```python
#Testing: red dot
r2 = r2_score(test_y, mymodel(test_x))
print(r2)
```

```python
#Predict
print(mymodel(2))
```

```python
#https://numpy.org/doc/stable/reference/generated/numpy.linspace.html  
plt.figure(figsize=(14,9))

plt.grid()
plt.scatter(train_x, train_y)
plt.scatter(test_x, test_y,color='r')

myline = numpy.linspace(1, 6, 100) 
#start. stop. no_of_sample +ve 
#Example: np.linspace(2.0, 3.0, num=5) == array([2.  , 2.25, 2.5 , 2.75, 3.  ])

plt.plot(myline, mymodel(myline))
plt.show()
```
