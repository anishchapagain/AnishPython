# Pie Chart 

```python
plt.pie(salary,labels=salary)
plt.legend()
#plt.grid(True)
plt.show()
```

shadow(True): adding 3D effect <br/>
startangle : Pie top angle to start with

```python
passAPlus = [70, 46, 79, 89, 62]
name= ['School A', 'School B', 'School C', 'School D', 'School E']

plt.pie(passAPlus, labels=name,shadow=True,startangle=90)

plt.title("School Grade Distribution - A+")
plt.show()
```

<b>explode:</b> explode/separate pie components <br/>
<b>autopct:</b> %value on pie component <br/>
<b>wedgeprops:</b> adding edgecolor<br />
<b>textprops:</b> text color<br/>
<b>colors:</b> component/wedge color<br/>

```python
plt.figure(figsize=(9,9))

passAPlus = [70, 46, 79, 89, 62]
name= ['School A', 'School B', 'School C', 'School D', 'School E']
explode=[0.1,0,0,0.1,0.09]

plt.pie(passAPlus, labels=name, explode=explode, shadow=True,startangle=70,
        autopct='%0.0f%%',
        wedgeprops={'edgecolor': 'black'},
        textprops=dict(color="r"), #{'color':'r'}
        colors=['r','b','g','y','k']
       )

plt.title("School Grade Distribution - A+")
plt.legend(passAPlus,title="APlus Grade",loc="upper left")

plt.tight_layout()
plt.show()

#TASKS: 20% 18% -> remove -> actual numerical value!
```
