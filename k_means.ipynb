# Task
Generate random data for age and income, apply K-means clustering, and visualize the results.

**K-means clustering** is an unsupervised machine learning algorithm used to partition a dataset into K distinct, non-overlapping subgroups or clusters. Here's a breakdown of how it works:

**Initialization**: You first specify the number of clusters (K) you want to create. The algorithm then randomly selects K data points from your dataset to serve as the initial centroids (the center points of each cluster).

**Assignment**: The algorithm then iterates through each data point in the dataset and assigns it to the cluster whose centroid is closest. This is typically measured using Euclidean distance.

**Update**: After all data points have been assigned to a cluster, the algorithm recalculates the position of each centroid based on the mean of all the data points assigned to that cluster.

**Iteration**: Steps 2 and 3 are repeated iteratively. With each iteration, the centroids shift their positions, and data points may be reassigned to different clusters based on the new centroid locations.

**Convergence**: The algorithm continues iterating until the centroids no longer significantly move or a maximum number of iterations is reached.

At this point, the clusters are considered stable, and the algorithm has converged.

## Generate random data

### Subtask:
Create a synthetic dataset for age and income using NumPy.


**Reasoning**:
Generate synthetic data for age and income using NumPy and combine them into a single array.



```python
import numpy as np

# Generate random data for age (100 integers between 20 and 60)
age = np.random.randint(20, 61, 100)

# Generate random data for income (100 floats between 30000 and 100000)
income = np.random.uniform(30000, 100001, 100)

# Combine age and income arrays into a single NumPy array with two columns
data = np.column_stack((age, income))

print(data[:5]) # Display the first 5 rows to verify
```

## Apply k-means clustering

### Subtask:
Use scikit-learn to apply K-means clustering to the generated data.


**Reasoning**:
Apply K-means clustering to the generated data using scikit-learn.



```python
from sklearn.cluster import KMeans

# Instantiate a KMeans object with n_clusters=3 and a random_state for reproducibility.
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# Fit the KMeans model to the generated data.
kmeans.fit(data)

# Store the cluster labels assigned to each data point.
labels = kmeans.labels_

print(labels[:5]) # Display the first 5 labels to verify
```

## Visualize the results

### Subtask:
Plot the clustered data to visualize the results.


**Reasoning**:
Plot the clustered data using matplotlib to visualize the results of the K-means clustering according to the instructions.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title("K-means Clustering of Age and Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```
